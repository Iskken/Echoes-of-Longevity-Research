# `take_2/` — network-first crawl

Core idea: build the forwarding graph first, then selectively crawl channels that had strong inbound edges. The hope was to reduce noise relative to the original seed-driven approach.

Outcome: the graph was small (~7k messages across a narrower channel set), and the channel-discovery rate did not justify the added bookkeeping complexity. Kept a few useful utility scripts — e.g. `checking-for-channel-existence.py` for validating channel handles before joining — but the main collection approach was abandoned.

Artifacts: `take_2/topic_modeling/bertopic_base.model/`, `take_2/messages.json` (small), `take_2/nodes.json`, `take_2/edges.json`.


# `take_3/` — abandoned early

Abandoned very early. The folder contains only `messages.json` (~794 KB) and the `discovered_channels_*.json` files. No topic model, no extended crawl. The reason is simply that work pivoted back to the `hybrid/` approach before `take_3/` could mature.


# Lessons that shaped the production `hybrid/` pipeline

1. **Seed-driven + forward discovery beats network-first** for a narrowly-scoped topic like longevity. The forwarding graph is useful as a *byproduct* of seed-driven crawling, not as the primary driver.
2. **Proxy routing is non-negotiable** — without DECODO SOCKS5 routing, rate-limiting makes bulk scraping impractical.
3. **Topic-relevance filtering at collection time** reduces downstream cleanup. Keywords are cheap; re-crawling is expensive.
4. **Keep `session.session` stable** — a single Telethon session file across runs is much more reliable than regenerating one per attempt.
