```python
def topic_diagnostics(df, topics_df, label):
    n_docs = len(df)
    n_topics = (topics_df['Topic'] != -1).sum()
    outliers = (df['topic'] == -1).sum()
    print(f"[{label}] docs={n_docs:,} | topics={n_topics} | outliers={outliers} ({outliers/n_docs:.2%})")
    display(topics_df.head())

topic_diagnostics(dfa, tpa, "SPECTER")
topic_diagnostics(dfm, tpm, "MiniLM")
```

**Search for the other unrelated topics**

```python
search_keywords = ["longevity", "healthy aging", "biohacking", "rejuvenation", "anti-aging", "nootropics"]
```

```python
def topic_is_relevant(rep, search_keywords):
    topic_words = eval(rep)  # convert string list to real list
    topic_words = [word.lower() for word in topic_words]
    return any(kw in word for word in topic_words for kw in search_keywords)

# Apply this to all topics
tpm['is_relevant'] = tpm['Representation'].apply(lambda rep: topic_is_relevant(rep, search_keywords))
```

```python
tpm[tpm['is_relevant']==False]
```

As you can see, I have filtered out unrelated topics based on the keywords. We will look at each cluster and decide if it should be deleted or left. In each topic, we will check if there might be related videos and how percentage are related. Based on that we will also manually look at them. If the topic is completely unrelated, then we will delete them.  

```python
#function to check if the video is related or unrelated

def is_video_related(text, keywords):
    text = str(text).lower()
    return any(kw in text for kw in keywords)
```

```python
#function to check how many percentage of the videos in the topics are related
def check_relevance_ratio(df, topic_id, keywords):
    df_topic = df[df['topic']==topic_id]
    relevant_count = df_topic['text'].apply(lambda x: is_video_related(x, keywords)).sum()
    return relevant_count/len(df_topic)
```

```python
keywords = ["longevity", "healthy aging", "biohacking", "rejuvenation", "anti-aging", "nootropics"]
```

```python
check_relevance_ratio(dfm, 4, keywords)
```

Let's assign their relevance ratio for each topic

```python
tpm['relevance_ratio'] = tpm['Topic'].apply(lambda t: check_relevance_ratio(dfm, t, keywords)) 
```

**Let's run the bertopic once again on the cleaned dataset**

```python
from bertopic import BERTopic
from sentence_transformers import SentenceTransformer
from sklearn.feature_extraction.text import CountVectorizer

model = SentenceTransformer("all-MiniLM-L6-v2") #for short texts

# Custom preprocessing with vectorizer
vectorizer_model = CountVectorizer(
    stop_words="english",  
    ngram_range=(1, 2),    
    max_df=0.85,           
    min_df=2,              
    max_features=4000  
)

# Creating BERTopic model
topic_model = BERTopic(
    embedding_model=model,
    vectorizer_model=vectorizer_model,
    nr_topics="auto",
    top_n_words=10,       
    calculate_probabilities=True,
    min_topic_size=30 #reduces overly small clusters
)
```

```python
texts = dfm['text'].fillna("").tolist()
```

```python
topics, probs = topic_model.fit_transform(texts)
```

```python
topic_model.get_topic_info()
```
